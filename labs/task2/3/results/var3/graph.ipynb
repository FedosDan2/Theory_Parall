```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
chunks = np.array([1, 4, 8, 16, 32, 64, 128, 256])
gain1 = np.array([7.29, 7.25, 6.46, 6.6, 6.65, 7.04, 4.58, 0.001]) 
gain2 = np.array([7.93, 8.22, 8.27, 8.31, 8.38, 8.3, 7.99, 0.002]) 

# Построение графиков
plt.plot(chunks, gain1,  'o-', color='red', linewidth=2, markersize=5, label='N = 750')
plt.plot(chunks, gain2,  'o-', color='blue', linewidth=2, markersize=5, label='N = 1000')

# Подписи и оформление
plt.xscale('log', base=2)
plt.xticks(chunks, labels=chunks)
plt.xlabel('Size of Chunk', fontsize=12)
plt.ylabel('Time(sec)', fontsize=12)
plt.title('Static', fontsize=14, fontweight='bold')
plt.legend(fontsize=11, loc='lower left')
plt.grid(True, alpha=0.3)

plt.show()
```

```python
chunks = np.array([1, 4, 8, 16, 32, 64, 128, 256])
gain1 = np.array([6.08, 5.4, 5.15, 5.37, 2.13, 0.44, 0.02, 0.006]) 
gain2 = np.array([8.36, 7.91, 7.59, 3.84, 2.0, 0.69, 0.04, 0.006]) 

# Построение графиков
plt.plot(chunks, gain1,  'o-', color='red', linewidth=2, markersize=5, label='N = 750')
plt.plot(chunks, gain2,  'o-', color='blue', linewidth=2, markersize=5, label='N = 1000')

# Подписи и оформление
plt.xscale('log', base=2)
plt.xticks(chunks, labels=chunks)
plt.yscale('log', base=2)
plt.yticks(gain1, labels=gain1)
plt.xlabel('Size of Chunk', fontsize=12)
plt.ylabel('Time(sec)', fontsize=12)
plt.title('Dynamic', fontsize=14, fontweight='bold')
plt.legend(fontsize=11, loc='lower left')
plt.grid(True, alpha=0.3)

plt.show()
```

```python
chunks = np.array([1, 4, 8, 16, 32, 64, 128, 256])
gain1 = np.array([4.94, 3.61, 1.69, 0.17, 0.05, 0.01, 0.005, 0.002]) 
gain2 = np.array([0.86, 0.09, 0.05, 0.02, 0.02, 0.01, 0.005, 0.007]) 

# Построение графиков
plt.plot(chunks, gain1,  'o-', color='red', linewidth=2, markersize=5, label='N = 750')
plt.plot(chunks, gain2,  'o-', color='blue', linewidth=2, markersize=5, label='N = 1000')

# Подписи и оформление
plt.xscale('log', base=2)
plt.xticks(chunks, labels=chunks)
plt.yscale('log', base=2)
plt.yticks(gain1, labels=gain1)
plt.xlabel('Size of Chunk', fontsize=12)
plt.ylabel('Time(sec)', fontsize=12)
plt.title('Guided', fontsize=14, fontweight='bold')
plt.legend(fontsize=11, loc='lower left')
plt.grid(True, alpha=0.3)

plt.show()
```

```python
chunks = np.array([1, 4, 8, 16, 32, 64, 128, 256])
times1 = np.array([7.29, 7.25, 6.46, 6.6, 6.65, 7.04, 4.58, 0.001]) 
times2 = np.array([6.08, 5.4, 5.15, 5.37, 2.13, 0.44, 0.02, 0.006]) 
times3 = np.array([4.94, 3.61, 1.69, 0.17, 0.05, 0.01, 0.005, 0.002]) 
avg = (times1 + times2 + times3) / 3

plt.figure(figsize=(10, 6))
static = plt.bar(chunks.astype(str), times1, color='steelblue', edgecolor='black', alpha=0.4, label="Static")
dynamic = plt.bar(chunks.astype(str), times2, color='green', edgecolor='black', alpha=0.4, label="Dynamic")
guided = plt.bar(chunks.astype(str), times3, color='red', edgecolor='black', alpha=0.4, label="Guided")

plt.title("N = 750, Different methods")
plt.yscale('log', base=2)
plt.yticks(avg, labels=avg)
plt.xlabel('Chunk Size', fontsize=12)
plt.ylabel('Time (sec)', fontsize=12)
plt.grid(axis='y', alpha=0.3, linestyle='--')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
