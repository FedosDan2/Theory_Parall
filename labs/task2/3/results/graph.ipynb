```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
threads = np.array([1, 2, 4, 6, 8, 16, 20, 40])
gain1 = np.array([1.000000, 2.050980, 3.464005, 4.491838, 4.717340, 5.743615, 5.734153, 3.708949]) 
gain2 = np.array([1.000000, 1.966490, 3.293156, 4.748082, 6.048553, 6.785851, 7.720751, 8.493660]) 
pret_ar = np.array([1.0, 2.0, 4.0, 6.0, 8.0, 16.0, 20.0, 40.0])

# Построение графиков
plt.plot(threads, pret_ar,  'o-', color='black', linewidth=2, markersize=5, label='Ideal Speedup (S = p)')
plt.plot(threads, gain1, 'o-', color='blue', linewidth=2, markersize=5, label='N = 750')
plt.plot(threads, gain2, 'o-', color='red', linewidth=2, markersize=5, label='N = 1000')

# Подписи и оформление
plt.xlabel('Number of Threads (p)', fontsize=12)
plt.ylabel('Speedup (S)', fontsize=12)
plt.title('Variant 2', fontsize=14, fontweight='bold')
plt.legend(fontsize=11, loc='upper left')
plt.grid(True, alpha=0.3)

for i, txt in enumerate(gain1):
    plt.annotate(f'{txt:.1f}', (threads[i], gain1[i]), 
                xytext=(5, -10), textcoords='offset points', fontsize=8, color='blue')
    
for i, txt in enumerate(gain2):
    plt.annotate(f'{txt:.1f}', (threads[i], gain2[i]), 
                xytext=(5, -20), textcoords='offset points', fontsize=8, color='red')

for i, txt in enumerate(pret_ar):
    plt.annotate(f'{txt:.1f}', (threads[i], pret_ar[i]), 
                xytext=(5, 20), textcoords='offset points', fontsize=8, color='black')


plt.show()
```

```python
chunks = np.array([8, 16, 32, 64, 128])
static = np.array([3.326017, 3.894719, 3.471683, 3.148852, 0.000425]) 
dynamic = np.array([0.300009, 0.255118, 0.010272, 0.008752, 0.000779]) 
guided = np.array([0.001062, 0.000640, 0.000550, 0.001858, 0.000376])

# Построение графиков
plt.plot(chunks, static,  'o-', color='black', linewidth=2, markersize=5, label='Static')
plt.plot(chunks, dynamic, 'o-', color='blue', linewidth=2, markersize=5, label='Dynamic')
plt.plot(chunks, guided, 'o-', color='red', linewidth=2, markersize=5, label='Guided')

# Подписи и оформление
plt.xlabel('Number of Chunks', fontsize=12)
plt.ylabel('Time', fontsize=12)
plt.title('Type of SChedule', fontsize=14, fontweight='bold')
plt.legend(fontsize=11, loc='upper left')
plt.grid(True, alpha=0.3)

for i, txt in enumerate(static):
    plt.annotate(f'{txt:.1f}', (chunks[i], static[i]), 
                xytext=(5, -10), textcoords='offset points', fontsize=8, color='blue')
    
for i, txt in enumerate(dynamic):
    plt.annotate(f'{txt:.1f}', (chunks[i], dynamic[i]), 
                xytext=(5, -20), textcoords='offset points', fontsize=8, color='red')

for i, txt in enumerate(guided):
    plt.annotate(f'{txt:.1f}', (chunks[i], guided[i]), 
                xytext=(5, 20), textcoords='offset points', fontsize=8, color='black')


plt.show()
```
