```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
threads = np.array([1, 2, 4, 6, 8, 16, 20, 40])
gain1 = np.array([1.000000, 1.948044, 3.578788, 4.258799, 3.688840, 5.200875, 5.452909, 6.013951]) 
gain2 = np.array([1.000000, 1.872239, 3.375784, 4.455898, 4.937491, 4.648109, 4.935366, 5.736945]) 
pret_ar = np.array([1.0, 2.0, 4.0, 6.0, 8.0, 16.0, 20.0, 40.0])

# Построение графиков
plt.plot(threads, pret_ar,  'o-', color='black', linewidth=2, markersize=5, label='Ideal Speedup (S = p)')
plt.plot(threads, gain1, 'o-', color='blue', linewidth=2, markersize=5, label='N = 750')
plt.plot(threads, gain2, 'o-', color='red', linewidth=2, markersize=5, label='N = 1000')

# Подписи и оформление
plt.xlabel('Number of Threads (p)', fontsize=12)
plt.ylabel('Speedup (S)', fontsize=12)
plt.title('Variant 1', fontsize=14, fontweight='bold')
plt.legend(fontsize=11, loc='upper left')
plt.grid(True, alpha=0.3)


plt.show()
```

```python
threads = np.array([1, 2, 4, 6, 8, 16, 20, 40])
gain1 = np.array([1.000000, 0.974022, 0.894697, 0.709800, 0.461105, 0.325055, 0.272645, 0.150349]) # 20000
gain2 = np.array([1.000000, 0.936120, 0.843946, 0.742650, 0.617186, 0.290507, 0.246768, 0.143424]) # 40000

# Построение графиков
plt.axhline(y=1, color='black', linestyle='--', label='Ideal')
plt.plot(threads, gain1, 'o-', color='blue', linewidth=2, markersize=5, label='750')
plt.plot(threads, gain2, 'o-', color='red', linewidth=2, markersize=5, label='1000')

# Подписи и оформление
plt.xlabel('Number of Threads (p)', fontsize=12)
plt.ylabel('Speedup (S / threads)', fontsize=12)
plt.title('Parallel Speedup Comparison', fontsize=14, fontweight='bold')
plt.legend(fontsize=11, loc='upper right')
plt.grid(True, alpha=0.3)

print(f"Best for 750: 2 threads - {max(gain1[1:])}")
print(f"Best for 1000: 2 threads - {max(gain2[1:])}")
plt.show()
```
