```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
threads = np.array([1, 2, 4, 6, 8, 16, 20, 40])
gain1 = np.array([1.000000, 1.993246, 3.972036, 5.902295, 7.767626, 13.740997, 15.457650, 23.685089]) # 20000
gain2 = np.array([1.000000, 1.948160, 3.712073, 5.439322, 7.077961, 11.373320, 13.834386, 23.969061]) # 40000
pret_ar = np.array([1.0, 2.0, 4.0, 6.0, 8.0, 16.0, 20.0, 40.0])

# Построение графиков
plt.plot(threads, pret_ar,  'o-', color='black', linewidth=2, markersize=5, label='Ideal Speedup (S = p)')
plt.plot(threads, gain1, 'o-', color='blue', linewidth=2, markersize=5, label='40000000')
plt.plot(threads, gain2, 'o-', color='red', linewidth=2, markersize=5, label='80000000')

# Подписи и оформление
plt.xlabel('Number of Threads (p)', fontsize=12)
plt.ylabel('Speedup (S)', fontsize=12)
plt.title('Parallel Speedup Comparison', fontsize=14, fontweight='bold')
plt.legend(fontsize=11, loc='upper left')
plt.grid(True, alpha=0.3)


plt.show()
```

```python
threads = np.array([1, 2, 4, 6, 8, 16, 20, 40])
gain1 = np.array([1.000000, 0.996623, 0.993009, 0.983716, 0.970953, 0.858812, 0.772883, 0.592127]) # 20000
gain2 = np.array([1.000000, 0.974080, 0.928018, 0.906554, 0.884745, 0.710832, 0.691719, 0.599227]) # 40000

# Построение графиков
plt.axhline(y=1, color='black', linestyle='--', label='Ideal')
plt.plot(threads, gain1, 'o-', color='blue', linewidth=2, markersize=5, label='20000')
plt.plot(threads, gain2, 'o-', color='red', linewidth=2, markersize=5, label='40000')

# Подписи и оформление
plt.xlabel('Number of Threads (p)', fontsize=12)
plt.ylabel('Speedup (S / threads)', fontsize=12)
plt.title('Parallel Speedup Comparison', fontsize=14, fontweight='bold')
plt.legend(fontsize=11, loc='upper right')
plt.grid(True, alpha=0.3)

print(f"Best for 20000: 6 threads - {max(gain1[1:])}")
print(f"Best for 40000: 6 threads - {max(gain2[1:])}")
plt.show()
```
