```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
threads = np.array([1, 2, 4, 6, 8, 16, 20, 40])
gain1 = np.array([1.0, 1.99, 3.94, 5.48, 6.74, 7.25, 8.5, 12.96]) # 20000
gain2 = np.array([1.0, 2.13, 4.16, 6.16, 8.03, 12.65, 11.42, 15.28]) # 40000
pret_ar = np.array([1.0, 2.0, 4.0, 6.0, 8.0, 16.0, 20.0, 40.0])

# Построение графиков
plt.plot(threads, pret_ar,  'o-', color='black', linewidth=2, markersize=5, label='Ideal Speedup (S = p)')
plt.plot(threads, gain1, 'o-', color='blue', linewidth=2, markersize=5, label='20000')
plt.plot(threads, gain2, 'o-', color='red', linewidth=2, markersize=5, label='40000')

# Подписи и оформление
plt.xlabel('Number of Threads (p)', fontsize=12)
plt.ylabel('Speedup (S)', fontsize=12)
plt.title('Parallel Speedup Comparison(STD::JTHREAD)', fontsize=14, fontweight='bold')
plt.legend(fontsize=11, loc='upper left')
plt.grid(True, alpha=0.3)


plt.show()
```

```python
threads = np.array([1, 2, 4, 6, 8, 16, 20, 40])
gain1 = np.array([1.000000, 0.997877, 0.987455, 0.914498, 0.842755, 0.453145, 0.425297, 0.324172]) # 20000
gain2 = np.array([1.000000, 1.068430, 1.041553, 1.027760, 1.004563, 0.791026, 0.571447, 0.382055]) # 40000

# Построение графиков
plt.axhline(y=1, color='black', linestyle='--', label='Ideal')
plt.plot(threads, gain1, 'o-', color='blue', linewidth=2, markersize=5, label='20000')
plt.plot(threads, gain2, 'o-', color='red', linewidth=2, markersize=5, label='40000')

# Подписи и оформление
plt.xlabel('Number of Threads (p)', fontsize=12)
plt.ylabel('Efficiency', fontsize=12)
plt.title('Parallel Speedup Comparison(STD::JTHEAD)', fontsize=14, fontweight='bold')
plt.legend(fontsize=11, loc='upper right')
plt.grid(True, alpha=0.3)

indices_gain1 = np.where(gain1 > 0.7)[0]
last_index1 = indices_gain1[-1]

indices_gain2 = np.where(gain2 > 0.7)[0]
last_index2 = indices_gain2[-1]

print(f"Best for 20000: {threads[last_index1]} threads - {gain1[last_index1]}")
print(f"Best for 40000: {threads[last_index2]} threads - {gain2[last_index2]}")
plt.show()
```
